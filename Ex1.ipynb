```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def gaussian(mean,cov,N=200):
    r1,r2 = np.random.default_rng().multivariate_normal(mean,cov,N).T
    return r1,r2
```

```python
s1a,s1b = gaussian([1,1],[[1,0],[0,1]])
s2a,s2b = gaussian([0,0],[[1,0.8],[0.8,1]])

plt.figure()
plt.scatter(s1a,s1b,label="s1")
plt.scatter(s2a,s2b,label="s2")
plt.grid()
plt.legend()
plt.show()
```

```python
def perceptron(x,w,b):
    wx = 0
    for i in range(len(w)):
        wx += w[i] * x[i]
    step = np.heaviside(wx-b,0.5)
    
    color = np.array([])
    for i in range(len(step)):
        if step[i] == 1:
            color = np.append(color,"b")
        else:
            color = np.append(color,"g")
            
    return step,color
```

```python
s1_test,col = perceptron(np.array([s1a,s1b]),np.array([0.2,0.4]),0.7)
print(s1_test)

plt.figure()
for i in range(len(s1_test)):
    plt.plot(s1a[i],s1b[i],"o",color=col[i])
plt.grid()
plt.show()
```

```python
s2_test,col2 = perceptron(np.array([s2a,s2b]),np.array([0.2,0.4]),0.7)
print(s2_test)

plt.figure()
for i in range(len(s2_test)):
    plt.plot(s2a[i],s2b[i],"o",color=col2[i])
plt.grid()
plt.show()
```
